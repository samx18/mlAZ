```python
# Importing the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing the dataset

dataset = pd.read_csv('Datasets/Position_Salaries.csv')
X = dataset.iloc[:,1:2].values # To get the features as a matrix instead of a vector
y = dataset.iloc[:,2].values


# Splitting the dataset into training set and the test set
#from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
# Fitting simple linear regression model to the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
# Fitting polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree= 2) # Create a ploynomial features with the specified degree ^
X_poly = poly_reg.fit_transform(X) # Fit & transform to create a ploynomial feature set
```

```python
X_poly
# Creates a 3 column matrix; 1st -> constant of ones,2nd -> X, 3rd -> squared based on degree
```

```python
# Create a polynomial linear regression model
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y) # Fit the polynomial feature set
```

```python
# Visualizing linear regression results
plt.scatter(X,y,color ='red')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()

```

```python
# Visualizing polynomial linear regression results
plt.scatter(X,y,color ='red')
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)),color='blue') # Use poly_reg.fit_transform(X) instead of just X_poly so the code can be used with a new feature of X  
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Ploynomial regression with 3 degress
# Fitting polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree= 3) # Create a ploynomial features with the specified degree ^
X_poly = poly_reg.fit_transform(X) # Fit & transform to create a ploynomial feature set
# Create a polynomial linear regression model
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y) # Fit the polynomial feature set
# Visualizing polynomial linear regression results
plt.scatter(X,y,color ='red')
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)),color='blue') # Use poly_reg.fit_transform(X) instead of just X_poly so the code can be used with a new feature of X  
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()

```

```python
# Ploynomial regression with 4 degress
# Fitting polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree= 4) # Create a ploynomial features with the specified degree ^
X_poly = poly_reg.fit_transform(X) # Fit & transform to create a ploynomial feature set
# Create a polynomial linear regression model
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y) # Fit the polynomial feature set
# Visualizing polynomial linear regression results
plt.scatter(X,y,color ='red')
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)),color='blue') # Use poly_reg.fit_transform(X) instead of just X_poly so the code can be used with a new feature of X  
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Creating a more fitted curve instead of a straight line between levels 

# Ploynomial regression with 4 degress
# Fitting polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree= 4) # Create a ploynomial features with the specified degree ^
X_poly = poly_reg.fit_transform(X) # Fit & transform to create a ploynomial feature set
# Create a polynomial linear regression model
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y) # Fit the polynomial feature set

# Creating a more fitted curve instead of a straight line between levels 
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid),1) # Reshape vector to a matrix


# Visualizing polynomial linear regression results
plt.scatter(X,y,color ='red')
plt.plot(X_grid,lin_reg_2.predict(poly_reg.fit_transform(X_grid)),color='blue') # Use poly_reg.fit_transform(X) instead of just X_poly so the code can be used with a new feature of X  
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
