```python
# Simple Linear Regresssion

# Importing the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing the dataset

dataset = pd.read_csv('Datasets/Salary_Data.csv')

X = dataset.iloc[:,:-1].values #matrix
y = dataset.iloc[:,1].values #vector

# Splitting the dataset into training set and the test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=1/3, random_state=0)

```

```python
# Fitting the simple linear regression model to the training set
from sklearn.linear_model import LinearRegression # import the LinearRegression class
regressor = LinearRegression()
regressor.fit(X_train,y_train) # Fit the regressor to the training set or the machine has learnt 
```

```python
# Predict test set results

y_pred = regressor.predict(X_test) # Predict based on the regressor value trained above using the test data 

# Creates a vector of the result
```

```python
# Visualizing the training results
import matplotlib.pyplot as plt
plt.scatter(X_train,y_train,color='red') # Create a scatter plot of the training dataset
plt.plot(X_train,regressor.predict(X_train),color ='blue') # We need to plot the predictions on the training data to get the best fit regression line
plt.title("Salary vs Expereince (Trainig Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()

# Real values will be in red (points) and the predicted values will be in blue (line)


```

```python
# Visualizing the test results
import matplotlib.pyplot as plt
plt.scatter(X_test,y_test,color='red') # Create a scatter plot of the test dataset
plt.plot(X_train,regressor.predict(X_train),color ='blue') # We do not need to change this as the regressor has already been trained
plt.title("Salary vs Expereince (Test Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()

# Real values will be in red (points) and the predicted values will be in blue (line)
```

```python

# To demonstrate there is no difference if you use the test set vs the train set for plotting the regression line

# Visualizing the test results
import matplotlib.pyplot as plt
plt.scatter(X_test,y_test,color='red') # Create a scatter plot of the test dataset
plt.plot(X_test,y_pred,color ='blue') # We do not need to change this as the regressor has already been trained
plt.title("Salary vs Expereince (Test Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()

# Real values will be in red (points) and the predicted values will be in blue (line)
```
