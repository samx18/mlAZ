```python
# Decision Tree Regression
# Importing the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing the dataset

dataset = pd.read_csv('Datasets/Position_Salaries.csv')
X = dataset.iloc[:,1:2].values # To get the features as a matrix instead of a vector
y = dataset.iloc[:,2].values

# Feature scaling

# from sklearn.preprocessing import StandardScaler
#sc_X = StandardScaler()
#sc_y = StandardScaler()
#X = sc_X.fit_transform(X)
#y = sc_y.fit_transform(y)


# Fitting the decision tree regression model to the dataset

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,y)




# Predicting
y_pred = regressor.predict([[6.5]])

# Decision Tree regression is non continuous ML model
# Use a higher resolution plot to visualize the results for a non continuous model



# Visualizing the Decision regression results
plt.scatter(X,y,color ='red')
plt.plot(X,regressor.predict(X),color='blue') # Use poly_reg.fit_transform(X) instead of just X_poly so the code can be used with a new feature of X  
plt.title('Truth or Bluff(Decission Tree Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()


```

```python

# Decision Tree regression is non continuous ML model
# Use a higher resolution plot to visualize the results for a non continuous model


X_grid = np.arange(min(X),max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color ='red')
plt.plot(X_grid,regressor.predict(X_grid),color='blue')
plt.title('Truth or Bluff(Decission Tree Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
